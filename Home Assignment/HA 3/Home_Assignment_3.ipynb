```python
import numpy as np
import matplotlib.pyplot as plt
```

### Define the sigmoid function

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

### Define the derivative of the sigmoid function

```python
def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))
```

### Generate x values

```python
x = np.linspace(-10, 10, 400) # Generate 400 points from -10 to 10 for a smooth curve
```

### Calculate y values for sigmoid and its derivative

y_sigmoid = sigmoid(x)
y_derivative = sigmoid_derivative(x)

```python
y_sigmoid = sigmoid(x)
y_derivative = sigmoid_derivative(x)
```

### Plotting

```python

plt.figure(figsize=(10, 6)) # Set the figure size
plt.plot(x, y_sigmoid, label='Sigmoid(x)', color='blue') # Plot sigmoid function in blue
plt.plot(x, y_derivative, label='Sigmoid Derivative', color='red', linestyle='--') # Plot derivative in red dashed line

# Add titles and labels
plt.title('Sigmoid Function and its Derivative')
plt.xlabel('x')
plt.ylabel('y')

# Add grid for better readability
plt.grid(True)

# Add legend to distinguish the functions
plt.legend()

# Show the plot
plt.show()
```
